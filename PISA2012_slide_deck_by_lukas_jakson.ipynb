### The Dataset

**What Is PISA?**

The Program for International Student Assessment (PISA) is a system of international assessments that allows countries to compare outcomes of learning as students near the end of compulsory schooling. PISA core assessments measure the performance of 15-year-old students in mathematics, science, and reading literacy every 3 years. Coordinated by the Organization for Economic Cooperation and Development (OECD), PISA was first implemented in 2000 in 32 countries.It has since grown to 65 education systems in 2012.

### Summary of findings

### Points scored and migration status in classic immigration countries compared with Germany

There is no or a non-significant difference between migrants and natives in the points scored, globally and in the classical immigration countries. In Germany, on the other hand, migrants score significantly worse than natives.

#### Differences in points scored in Germany depending on migrants' country of immigration

Most of the 21.2 million people in Germany with a migration background in 2019 were from Turkey (13.3 percent), followed by Poland (10.5 percent) and former Soviet Union (6.5 percent).



The samples are probably too small to draw conclusions, however, Polish-born students seem to do better than those of Russian and Turkish origin. 

#### Differences of migrants of Turkish origin in the scores achieved in the PISA study in Germany depending on gender.

In recent years, there have been frequent reports in the German media that students of Turkish origin are among the weakest at school. There have also been reports that girls of Turkish origin, if allowed by their parents, have caught up in school performance and overtaken male students of Turkish origin.



Again, the sample is too small to draw a meaningful conclusion. However, one can see that there is something to the media reports. Especially in reading, girls of Turkish origin are better than boys - on average 10.5 %.
